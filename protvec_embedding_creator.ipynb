```python
import biovec
import timeit
import pandas as pd
import numpy as np
from biovec import utils
import utils

#specifying path to model weights
model_path = "./trained_models/swissprot-reviewed-protvec.model"

#defining paths for incoming sequence and outcoming embedding files
incoming_panda = "protabank.csv"
outcoming_panda = "protvec.tsv"
```

```python
#loading the model
model = biovec.models.load_protvec(model_path)

#loading incoming dataframe 
seq_dictionary = pd.read_csv(incoming_panda, sep=",",header=0)
seq_dictionary.head(5)
```

```python
def extract_embedings (seq):

    try:
        pred = model.to_vecs(seq)
        pred = pred.sum(axis=0)

    except:
        #catch exception
        pred = "Failed"
        print("failed")
        
    return pred


def data_df(protein_name,seq):
    # Forward
    protein_embedding = extract_embedings(sequence)
    
    if protein_embedding == "Failed":
        #catch failed cases
        embedding_dict = {'protein_name': protein_name, 'embedding': np.nan}
        
    else:

        #create dic with embeddings
        embedding_dict = {'protein_name': protein_name, 'embedding': [list(protein_embedding)]}
    
    #create a df with single protein embeddings
    df = pd.DataFrame(embedding_dict, index = [0])
    return df
```

```python
%%time
df_embeddings = []
#iterate through dataframe

for index, row in seq_dictionary.iterrows():
    sequence_name = index
    sequence = row[0]

    print("Iteration: ",index,"Sequence name: ",sequence_name)
    
    #extract embedding as dataframe and append to dataframe list
    df_embeddings.append(data_df(sequence_name,sequence))
```

```python
df_embeddings[0].head()
```

```python
#concatenating all single dataframes into one dataframe
df_embed = pd.concat(df_embeddings)
df_embed.index= list(range(len(df_embeddings)))
df_embed.head(5)
```

```python
#validate that output from dataframe can be converted to vector
print(len(df_embed.loc[0,'embedding']))
```

```python
result = pd.concat([seq_dictionary,df_embed],axis = 1)
```

```python
result.to_csv("protvec_concat.csv",index=False)
```
